# Theoretical Questions

Ans 1: Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves testing an assumption (hypothesis) about a population parameter (like the mean or proportion) using a sample.

Ans2: Null Hypothesis (H₀): The default or initial assumption that there is no effect or no difference (e.g., the population mean equals a specific value).

Alternative Hypothesis (H₁ or Ha): The competing claim that there is an effect or a difference (e.g., the population mean is not equal to a specific value).

They are mutually exclusive — only one can be true based on the data.



Ans3: The significance level (α) is the threshold for rejecting the null hypothesis. Common values are 0.05, 0.01, or 0.10.

It represents the probability of making a Type 1 error (rejecting a true null hypothesis).

It's important because it defines the boundary for deciding whether your test result is statistically significant.

Ans4: A P-value is the probability of obtaining results at least as extreme as the observed results, assuming the null hypothesis is true.



Ans5: If P-value ≤ α: Reject the null hypothesis — the result is statistically significant.

If P-value > α: Do not reject the null hypothesis — not statistically significant.
A smaller P-value indicates stronger evidence against the null hyp

Ans6: Type 1 Error (False Positive): Rejecting a true null hypothesis.

Probability = α

Type 2 Error (False Negative): Not rejecting a false null hypothesis.

Probability = β

Ans7: One-tailed test: Tests for the possibility of an effect in one direction (e.g., greater than or less than a value).

Two-tailed test: Tests for an effect in both directions (e.g., not equal to a value).

Choice depends on the research question.



Ans8: A Z-test is used when:

The population variance is known.

The sample size is large (typically n > 30).
It compares the sample mean to the population mean to determine if there is a significant difference.

Ans10: The T-distribution is similar to the normal distribution but has heavier tails.
Use it when:

The sample size is small (n < 30)

# Z-test

Used --	Large sample, known population σ

Distribution -- Standard Normal (Z)

Tail behavior -- Narrower tails

# T-test

Used --	 Small sample, unknown population σ

Distribution -- Student’s T-distribution

Tail behavior -- Heavier tails (more uncertainty)

Ana12: A T-test compares means to determine if they are significantly different.
Types include:

One-sample T-test: Compares sample mean to population mean.

Two-sample T-test: Compares means of two independent samples.

Paired T-test: Compares means from the same group at different times

Ans13: Both are used to test hypotheses about means. The T-test is essentially a Z-test with adjusted uncertainty due to smaller sample sizes or unknown variance. As sample size increases, the T-distribution approaches the normal (Z) distribution.



Ans14: A confidence interval (CI) gives a range of values within which a population parameter likely falls, based on a sample.
Example: A 95% CI means there's a 95% chance the interval contains the true mean.
If a CI does not include the null value (e.g., 0 for mean difference), it's statistically significant.

Ans 19: The F-distribution compares variances. It's used when testing:

Variance ratios (e.g., comparing spread of two datasets)

Ans20: ANOVA (Analysis of Variance) tests whether the means of 3 or more groups are significantly different.

Assumptions:

Independence of observations

Normal distribution of residuals

Equal variances (homogeneity of variance)

Ans21: One-way ANOVA: Tests one independent variable (e.g., effect of diet on weight loss).

Two-way ANOVA: Tests two independent variables and their interaction (e.g., diet and exercise).

Repeated Measures ANOVA: Used for correlated data (e.g., same subjects measured over time).

```python
#Ans1
from scipy.stats import norm

def z_test(sample_mean, population_mean, std_dev, n):
    z = (sample_mean - population_mean) / (std_dev / (n ** 0.5))
    p_value = 2 * (1 - norm.cdf(abs(z)))
    return z, p_value

# Example
z, p = z_test(52, 50, 5, 30)
print(f"Z = {z:.2f}, P-value = {p:.4f}")
```

```python
#Ans2
import numpy as np
sample = np.random.normal(50, 10, 100)
z, p = z_test(np.mean(sample), 50, 10, len(sample))
print(f"Simulated Z = {z:.2f}, P-value = {p:.4f}")
```

```python
#Ans3
def one_sample_z_test(data, population_mean, population_std):
    sample_mean = np.mean(data)
    n = len(data)
    return z_test(sample_mean, population_mean, population_std, n)

# Example
data = np.random.normal(50, 10, 100)
z, p = one_sample_z_test(data, 50, 10)
print(f"Z = {z:.2f}, P-value = {p:.4f}")
```

```python
#Ans4
import matplotlib.pyplot as plt

z_val = 2.1
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, label='Standard Normal')
plt.axvline(-1.96, color='red', linestyle='--', label='Critical Values')
plt.axvline(1.96, color='red', linestyle='--')
plt.axvline(z_val, color='green', linestyle='-', label='Z-value')
plt.fill_between(x, y, where=(x <= -1.96) | (x >= 1.96), color='red', alpha=0.3)
plt.title('Two-Tailed Z-Test')
plt.legend()
plt.grid()
plt.show()
```

```python
#Ans5
def visualize_errors(mu0=0, mu1=1, std=1, n=30, alpha=0.05):
    se = std / np.sqrt(n)
    crit_val = norm.ppf(1 - alpha / 2)
    x = np.linspace(-4, 4, 1000)

    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    plt.plot(x, y0, label='H0', color='blue')
    plt.plot(x, y1, label='H1', color='green')

    plt.axvline(mu0 + crit_val * se, color='red', linestyle='--', label='Critical Region')
    plt.axvline(mu0 - crit_val * se, color='red', linestyle='--')

    plt.fill_between(x, y0, where=(x <= mu0 - crit_val * se) | (x >= mu0 + crit_val * se), color='blue', alpha=0.3, label='Type I Error')
    plt.fill_between(x, y1, where=(x > mu0 - crit_val * se) & (x < mu0 + crit_val * se), color='green', alpha=0.3, label='Type II Error')

    plt.title('Type I and Type II Errors')
    plt.legend()
    plt.grid()
    plt.show()

visualize_errors()
```

```python
#Ans9
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    se = stats.sem(data)
    n = len(data)
    ci = t.interval(confidence, n-1, loc=mean, scale=se)
    return ci

data = np.random.normal(100, 10, 50)
ci = confidence_interval(data)
print("95% Confidence Interval:", ci)
```

```python
#Ans10
def margin_of_error(data, confidence=0.95):
    se = stats.sem(data)
    n = len(data)
    t_crit = t.ppf((1 + confidence) / 2, n - 1)
    return t_crit * se

moe = margin_of_error(data)
print("Margin of Error:", moe)
```

```python
#Ans11
def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):
    p_not_a = 1 - p_a
    p_b = p_b_given_a * p_a + p_b_given_not_a * p_not_a
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b

# Example
posterior = bayes_theorem(0.01, 0.9, 0.1)
print("P(A|B):", posterior)
```

```python
#Ans13
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Observed data in contingency table form
observed = np.array([[30, 10],
                     [20, 40]])

# Perform Chi-square test and get expected frequencies
chi2, p, dof, expected = chi2_contingency(observed)

# Display results
print("Observed Frequencies:")
print(pd.DataFrame(observed))

print("\nExpected Frequencies:")
print(pd.DataFrame(np.round(expected, 2)))

print(f"\nChi-square Statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")

```

```python
#Ans14
from scipy.stats import chisquare

# Example: observed vs. expected frequencies
observed = np.array([18, 22, 20, 25, 15])  # Sample counts
expected = np.array([20, 20, 20, 20, 20])  # Uniform distribution

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Observed:", observed)
print("Expected:", expected)
print(f"\nChi-square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")

```
